##1. Cargar los datos del csv en un pandas dataframe.

```python
import pandas as pd

covid19 = pd.read_csv('https://raw.githubusercontent.com/felipeaninir/random/main/country_vaccinations.csv')
covid19
```

2. Para cada país indicar la fecha que empezó a vacunar.

```python
not_nan = covid19.dropna(axis=0, subset=['daily_vaccinations_raw']) #Tome la fecha donde: daily_vaccinations_raw != NaN
país_fecha_inicio_vacunacion = not_nan.groupby(["country"]).head(1)
©, len(covid19["country"].unique()))
print('\nFecha que empezó a vacunar cada pais\n', país_fecha_inicio_vacunacion.iloc[:, [0,2]]) #91 de 129 paises registrados llevan vacunados a la fecha del archivo
```

**3. Ordenar país-fecha_inicio_vacunación, de acuerdo a fecha en orden**
**creciente (más antiguos primero).**

```python
print('\nFechas en orden creciente de inicio vacunados\n', país_fecha_inicio_vacunacion.sort_values("date").iloc[:, [0,2]])
```

**4. Indicar la fecha del último registro para cada país**

```python
ultimo_registro = not_nan.groupby(["country"]).tail(1)
print('\nFechas ultimo registro cada pais\n', ultimo_registro.iloc[:, [0,2]])
```

**5. Filtrar el dataframe para solo considerar los datos de Chile**

```python
chile = covid19[covid19["country"] == "Chile"]
chile
```

**6. Exportar el dataframe con los datos de Chile a un archivo Excel(.xlsx)**

```python
chile.to_excel("chile.xlsx", sheet_name="chile")
```
